```python
import json

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
with open('./data_A4_2019-11-26.json', 'r') as f:
    data = json.load(f)
```

```python
flow = np.array(data['data']['flow'])[:100,:]
speed = np.array(data['data']['speed'])[:100,:]
density = flow / speed
```

```python
plt.scatter(density, flow, marker='+', s=10)
plt.xlabel('Density')
plt.ylabel('Flow')
plt.title('Density vs Flow')
```

```python
plt.scatter(flow, speed, marker='+', s=10)
plt.ylim([0, 130])
plt.xlabel('Flow')
plt.ylabel('Speed')
```
